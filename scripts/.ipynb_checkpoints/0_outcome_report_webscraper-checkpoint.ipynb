```python
###script to scrape outcome reports from nsf reporter
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
##import the list of IDs
#this csv was downloaded from nsf reporter:
# 26 duplicate grants removed
nsf_df = pd.read_csv ('../input/grants_igert_nrt.csv',encoding='latin-1')

## set up new datafame with columns for all the info we want
df2 = pd.DataFrame(columns=['id','url','outcomes'])
df3 = pd.DataFrame(columns=['id','url','outcomes'])

#make a list of all the award ids to convert into urls
url_list = nsf_df.iloc[:,0].tolist()
```

```python
#iterate through all the urls
for k in range(len(url_list)):
    url_end = str(url_list[k])
    url = 'https://www.nsf.gov/awardsearch/showAward?AWD_ID='+url_end

    #use soup to grab all the paragraph tags then put them into a big string
    page = requests.get (url)
    soup = BeautifulSoup(page.content, 'html.parser')
    para_list = soup.find_all('p')
    print (para_list)
    para_string = ""
    for p in para_list:
        para_string += str(p)
    print (para_string)
    # pull out a single instance of the project outcomes report from the big string
    ##define the regex for pulling out the project outcomes report
    regex= 'PROJECT OUTCOMES REPORT((.|\n)*?)Modified by:.*'
    try:
        por_match = re.search(regex, para_string)
        por_string = (por_match.group())
    except AttributeError:
        por_string = ""

    #add the outcome report string to the master df
    df3.loc[1] = [url_end,url,por_string]
    df2 = pd.concat([df2, df3], ignore_index=True)
```

```python
##clean up html formatting
def remove_html_bs4(text):
    if isinstance(text, str): # Ensure the input is a string
        soup = BeautifulSoup(text, 'html.parser')
        return soup.get_text(separator=' ', strip=True) # Use separator and strip for cleaner output
    return text # Return original if not a string

# Apply the function to the column
df2['Cleaned Outcome Report'] = df2['outcomes'].apply(remove_html_bs4)
```

```python
## remove introductory phrasing and extra spaces
phrases = ['PROJECT OUTCOMES REPORT Disclaimer This Project Outcomes Report for the General Public is displayed verbatim as submitted by the',
           'Principal Investigator \(PI\) for this award.', 
           'Any opinions, findings, and conclusions or',
           'recommendations expressed in this Report are those of the PI and do not necessarily reflect the',
           'views of the National Science Foundation; NSF has not approved or endorsed its content.',
            '\n',
            '   ']

# Apply the regex to the column
for i in phrases:
    df2['Cleaned Outcome Report'] = df2['Cleaned Outcome Report'].str.replace(i, '', regex=True)
```

```python
##attach outcome reports to grant info
#make the id numbers strings to match dfs
nsf_df['id'] = nsf_df['AwardNumber'].astype(str)
new_df = pd.merge(nsf_df, df2, on='id')

#drop unneeded columns
new_df = new_df.drop(['id', 'url','outcomes'], axis=1)
#rename
new_df = new_df.rename(columns={'Cleaned Outcome Report':'Outcome Report'}
```

```python
#save the final df to csv
new_df.to_csv('../output/grants_0.csv', encoding = 'utf-8', index=False)
```
